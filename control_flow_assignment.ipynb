```python
#Task 1: Discount Rules (if / elif / else)
#2. Apply the following discount rules and print the final amount using (if / elif / else):
# order_amount >= 2000 → 15% discount
# 1500 <= order_amount < 2000 → 10% discount
# 1000 <= order_amount < 1500 → 7% discount
# otherwise → 0% discount
order_amount = int(input("Enter the order amount: "))
if order_amount >= 2000:
    discount = 0.15
elif order_amount >= 1500 and order_amount < 2000:
    discount = 0.10
elif order_amount >= 1000 and order_amount < 1500:
    discount = 0.07
else:
    discount = 0

final_amount = order_amount - (order_amount * discount)
print("The final amount after applying the discount is:", final_amount)
```

```python
#Task 2: Process Multiple Orders (for loop)
#1. Given a list of order amounts: orders = [1200, 2500, 800, 1750, 3000], use a for loop
#to apply the discount rules from Task 1 to each order and print a summary table
#showing: order_amount -> discount% -> final_amount.
orders = [1200, 2500, 800, 1750, 3000]
print("order_amount -> discount% -> final_amount")
for order in orders:
    if order >= 2000:
        discount = 15
    elif order >= 1500 and order < 2000:
        discount = 10
    elif order >= 1000 and order < 1500:
        discount = 7
    else:
        discount = 0
    final_amount = order - (order * discount / 100)
    print(f"{order} -> {discount}% -> {final_amount}")
```

```python
#Task 2: Process Multiple Orders (for loop)
#2. Also compute and print the total revenue after discounts.
orders = [1200, 2500, 800, 1750, 3000]
total_revenue = 0
print("order_amount -> discount% -> final_amount")
for order in orders:
    if order >= 2000:
        discount = 15
    elif order >= 1500 and order < 2000:
        discount = 10
    elif order >= 1000 and order < 1500:
        discount = 7
    else:
        discount = 0
    final_amount = order - (order * discount / 100)
    total_revenue += final_amount
    print(f"{order} -> {discount}% -> {final_amount}")
print(f"Total revenue after discounts: {total_revenue}")
```

```python
#Task 3: User Menu (while loop + break/continue)
#1. Implement a simple while-loop driven menu that keeps asking the user to choose an
#action until they press q to quit.
#Menu options:
#1 - Add order amount to a running list (use input())
#2 - Show all orders and totals after applying discounts
#q - Quit
orders = []
while True:
    choice = input("Choose an option (1 - Add order, 2 - Show orders, q - Quit): ")
    if choice == '1':
        try:
            order_amount = int(input("Enter the order amount: "))
            orders.append(order_amount)
        except ValueError:
            print("Error: Please enter a valid integer.")
    elif choice == '2':
        total_revenue = 0
        print("order_amount -> discount% -> final_amount")
        for order in orders:
            if order >= 2000:
                discount = 15
            elif order >= 1500 and order < 2000:
                discount = 10
            elif order >= 1000 and order < 1500:
                discount = 7
            else:
                discount = 0
            final_amount = order - (order * discount / 100)
            total_revenue += final_amount
            print(f"{order} -> {discount}% -> {final_amount}")
        print(f"Total revenue after discounts: {total_revenue}")
    elif choice == 'q':
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid option. Please choose again.")
```

```python
#Task 4: Loop Control with Conditions (break & continue)
#1.Given a list of daily sales: daily = [200, 150, 0, 400, 50, -1, 300], iterate through it with
#a for loop and:
#o If a day's value is -1, treat it as corrupted data and break the loop (stop processing).
#o If a day's value is 0, treat it as a day with no sales and continue (skip adding to
#revenue).
#• For valid positive sales, add to total_sales and print the running total.
daily = [200, 150, 0, 400, 50, -1, 300]
total_sales = 0
for day in daily:
    if day == -1:
        print("Corrupted data found. Stopping processing.")
        break
    elif day == 0:
        print("No sales for this day. Skipping.")
        continue
    else:
        total_sales += day
        print(f"Added {day} to total sales. Running total: {total_sales}")
```

```python
#Task 4: Loop Control with Conditions (break & continue)
#2. Print the final total after the loop completes (or stops due to corrupted data).
print(f"Final total sales: {total_sales}")
```
