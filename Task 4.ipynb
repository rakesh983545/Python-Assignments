```python
#1. Using the products list and price_dict, create a list of tuples named catalog where each tuple is (product_name, price, category).
products = ["Laptop", "Smartphone", "Tablet", "Headphones", "Smartwatch", "Camera"]
categories = ["Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Electronics"]
price_dict = {
    "Laptop": 999,
    "Smartphone": 699,
    "Tablet": 399.99,
    "Headphones": 149.99,
    "Smartwatch": 249.99,
    "Camera": 799.99
}
catalog = [(products[i], price_dict[products[i]], categories[i]) for i in range(len(products))]
print("Catalog:", catalog)
```

```python
#2. From catalog, create a new dictionary category_to_products that maps each category to a list of product names in that category.
category_to_products = {}
for product, price, category in catalog:
    if category not in category_to_products:
        category_to_products[category] = []
    category_to_products[category].append(product)
print("Category to Products Mapping:", category_to_products)
```

```python
#3. Print all products that belong to the category that has the maximum number of Products.
max_products = max(len(products_list) for products_list in category_to_products.values())
max_category = [category for category, products_list in category_to_products.items() if len(products_list) == max_products][0]
print(f"Products in the category with maximum number of products ({max_products} products): {category_to_products[max_category]}")
```
